```python
# Dealing with takehome_users data
takehome_users.creation_time = pd.to_datetime(takehome_users.creation_time) #Setting the column in their respectable datetime format
takehome_users.last_session_creation_time = pd.to_datetime(takehome_users.last_session_creation_time)
takehome_users.drop(['name','email'],axis=1,inplace=True) #Removing unnecessary data
takehome_users.rename(columns={'object_id':'user_id'},inplace=True) # Matching same column value of both file with same column name

# Dealing with takehome_user_engagement data
takehome_user_engagement.time_stamp = pd.to_datetime(takehome_user_engagement.time_stamp) 
takehome_user_engagement["year"]=takehome_user_engagement.time_stamp.dt.year
takehome_user_engagement["month"]=takehome_user_engagement.time_stamp.dt.month
takehome_user_engagement["day"]=takehome_user_engagement.time_stamp.dt.day
takehome_user_engagement["week"]=takehome_user_engagement.time_stamp.dt.week
takehome_user_engagement['year_week'] = list(zip(takehome_user_engagement.year,takehome_user_engagement.week))
```
